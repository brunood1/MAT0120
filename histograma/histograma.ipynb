```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
pgm = []

with open('wolverine.pgm', 'r') as file: 
    lines = file.readlines()
    
for line in lines: 
    if line[0] == "#": 
        continue 
    elif line[0] == "P": 
        P = line[1] 
    else:
        try:
            space = line.index(' ')
            l = int(line[0:space])
            h = int(line[space+1:])
        except:
            pgm.append(int(line.strip()))
```

```python
plt.hist(pgm, bins=51, color='grey')
plt.savefig('histograma_pgm.pdf')
plt.show()
```

```python
ppm = []
RGB = []

with open('wolverine.ppm', 'r') as file:
    lines = file.readlines()
    
mod = 0
for line in lines:
    if line[0] == "#":
        continue
    elif line[0] == "P":
        P = line[1] 
    else:
        try:
            space = line.index(' ')
            l = int(line[0:space])
            h = int(line[space+1:])
        except:
            ppm.append(int(line.strip()))
            RGB.append(mod%3)
            mod = mod + 1
```

```python
RED = []
GREEN = []
BLUE = []
for i in range(len(ppm)):
    if   RGB[i] == 2:
        RED.append(ppm[i])
    elif RGB[i] == 0:
        GREEN.append(ppm[i])
    elif RGB[i] == 1:
        BLUE.append(ppm[i])
        
GREEN.remove(GREEN[0])
```

```python
fig, axs = plt.subplots(3, 1, figsize=(10,10))

axs[0].hist(RED, bins=51, color='red')
axs[1].hist(GREEN, bins=51, color='green')
axs[2].hist(BLUE, bins=51, color='blue')
plt.savefig('histograma_ppm.pdf')
plt.show()
```
